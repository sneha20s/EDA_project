**inplace**
- when we do the operation the result output is not overwriting
- means the original dataframe is same as it is
- in order to do this we need to use inplace 
- by default inplace=False
- make **inplace=True** if you want overwrite the reult

```python
d={'Names':['Ramesh','Suresh','Sathish'],
   'Age':[20,22,24],
   'City':['Hyd','Pune','Blr'],
  'Marks':[34,56,54]}
df=pd.DataFrame(d)
df
```

```python
df.drop('Marks',axis=1,inplace=True)
df
```
