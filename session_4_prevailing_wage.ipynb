## Winsoriazation

- whatever the outliers less than lb fill with lb
- whatever the outliers greater than ub fill with ub
- in numpy we have method **clip**


```python
step-1: get the wage data values
step-2: calculate lb
step-3: calculate ub
step-4: check the parameters and apply in numpy np.clip()
step-5: again implement above visualization analysis
```

```python

wage_data = visa_df['prevailing_wage'].values
q1 = np.percentile(wage_data, 25)
q3 = np.percentile(wage_data, 75)
iqr = q3 - q1
lb = q1 - 1.5 * iqr    # lower bound
ub = q3 + 1.5 * iqr    # upper bound
wage_data_clipped = np.clip(wage_data, lb, ub)

# Step-5: Visualization â€“ Boxplot + Histogram
plt.figure(figsize=(10,6))
plt.subplot(2,2,1).hist(wage_data)
plt.subplot(2,2,2).hist(wage_data_clipped)
plt.subplot(2,2,3).boxplot(wage_data,vert=False)
plt.subplot(2,2,4).boxplot(wage_clipped,vert=False)
plt.show()
# Boxplot

```
