```python
folder_name='Frequency_table'
os.makedirs('Frequency_table',exist_ok=True)
#step-2: make a new path
directory=os.getcwd()
new_dir=os.path.join(directory,'Frequency_table')
#step-3 create a dataframe
for i in cat [1:]:
    keys=visa_df[i].value_counts().keys()
    values=visa_df[i].value_counts().values
    df=pd.DataFrame(zip(keys,values),
                    columns=['class', 'frequency'])
#step-4: make the new path
    file_name=f'{i}_df.csv'
    path=os.path.join(new_dir,f'{i}_df.csv')
    #step-5:save into dataframe
    df.to_csv(path,index=False)
```

```python
folder_name='pie_chart'
os.makedirs('pie_chart',exist_ok=True)
directory=os.getcwd()
new_dir=os.path.join(directory,'pie_chart')
for i in cat[1:]:
    counts = visa_df[i].value_counts()
    keys=visa_df[i].value_counts().keys()
    values=visa_df[i].value_counts() 
    plt.pie(values, labels=keys, autopct='%1.2f%%')
    plt.title(i)  
    plt.show()
    file_name=f'{i}_df.csv'
    path=os.path.join(new_dir,f'{i}_df.csv')
    df.to_csv(path,index=False)
```

```python
folder_name='bar_chart'
os.makedirs('bar_chart',exist_ok=True)
directory=os.getcwd()
new_dir=os.path.join(directory,'bar_chart')
for i in cat[1:]:
    keys=visa_df[i].value_counts().keys()
    values=visa_df[i].value_counts()
    plt.figure(figsize=(10,4))
    plt.bar(keys,values)
    plt.xlabel('classes')
    plt.ylabel('Frequency')
    plt.title(f'{i}_barchart')
    file_name=f'{i}_df.csv'
    path=os.path.join(new_dir,f'{i}_df.csv')
    df.to_csv(path,index=False)
```
